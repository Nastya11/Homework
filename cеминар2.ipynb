```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

def p(v, t):
    v = np.asarray(v)
    res =  (-3/v**2) + (8/3)*t/(v- 1/3)
    return res
```

```python
%matplotlib notebook
t=0.88
v = np.linspace(0.5,5,100)
plt.plot(v,p(v,t))
plt.grid()
```

```python
def solve_eq(p,t):
    matrix = np.array([[0,0,1/p], [1,0,-3/p], [0,1,(8*t + p)/(3*p)]])
    eig = np.linalg.eig(matrix)[0]
    return eig
```

```python
p=0.4
a = solve_eq(p, t)
b = np.roots([p, (-p/3 - 8*t/3), 3, -1])
print(a, '\n', b)
```
