# Метод парабол

```python
import math
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
eps = 1e-10
a = 4.5
c = 6.5
b = (a+c)/2.0
```

```python
def f(x): 
    '''наша функция -cosx'''
    f=-(math.cos(x))
    return f
```

```python
def p_extr(f, a, c): 
    '''Экстремум параболы'''
    b = (a+c)/2.0
    x = b + 1.0/2.0 * ( (c-b)**2 * (f(a)-f(b)) - (b-a)**2 * (f(c)-f(b)))/((c-b) * (f(a)-f(b)) + (b-a) * (f(c)-f(b)))
    return x 
```

```python
def min_pmethod(f, a, c):
    '''метод парабол'''
    b = (a+c)/2.0
    minx = p_extr(f, a, c)
    if abs(f(minx)-f(b)) >= eps:  
        if f(minx)<f(b) :
        
            if minx < b:
                a = a
                c = b
                return p_extr(f, a, c)
        
            elif minx > b :
                a = b
                c = c
                return p_extr (f, a, c)
            
        else :
            return (a+c)/2.0
    else:
        return (a+c)/2.0
```

Найдем производную: (-cosx)'=sinx. Условие экстремума: f'(x)=0 => x=Пn, n принадлежит целым числам. Возьмем промежуток, где находится 1 минимум, например, [4.5,6.5]. Минимум будет в точке х=2П = 6,2831853... Посмотрим, какой ответ получится нашим методом:

```python
x=min_pmethod(f, a, c)
print('min x =', x, 'min y = ', f(x))
```

```python
%matplotlib notebook
xx = np.arange(a,c, 0.002)
minimum_1 = min_pmethod(f,a,c)
plt.plot(xx, [f(el) for el in xx], '-')
plt.plot(xx, [f(minimum_1) for el in xx], '-')
```
