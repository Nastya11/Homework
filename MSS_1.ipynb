```python
import numpy as np
from scipy.optimize import root
import matplotlib.pyplot as plt
```

```python
def func_c(v, r, r0, v0):
    return v**2/2+ np.log((v0*r0**2)/(v*r**2)) - 1/r + 1/r0 - v0**2/2
```

```python
def solution(r0, v0, r_min, r_max, n):
    for i in range(n):
        r = np.linspace(r_min, r_max, n)
        crange = range(0, n)
        v = [root(func_c, r_min, args=(r[ci],r0,v0[i])).x[0] for ci in crange]
        plt.plot(r,v)
    plt.show()
    return r,v
```

```python
r0=0.5
v0 = np.linspace(0,10,100)
r_min =0.01
r_max = 1
n = 50
sol = solution(r0, v0, r_min, r_max, n)
```
