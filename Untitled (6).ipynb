```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x,a,b,c):
    arg=(x-a)/b
    return c*np.exp(-(arg)**2)
```

```python
xa, xb= -2.,2.
npts=20
```

```python
xx = np.linspace(xa,xb)
```

```python
np.random.seed(123)
```

```python
noise=np.random.normal(loc=0, scale=0.03, size=xx.size)
```

```python
a=0.5
b=1
c=2
```

```python
yy = f(xx, a, b, c)
```

```python
noisy_y=yy + 2*noise
%matplotlib notebook
plt.plot(xx,yy,'--')
plt.plot(xx, noisy_y, 'o')
```

```python
from scipy.optimize import curve_fit
```

```python
curve_fit(f, xx, noisy_y, [0.5,1.,0.2])
```

```python
def fl(x,a,b,c):
    return a*x**2+b*x+c
```

```python
mask=noisy_y>0
```

```python
popt, pcov = curve_fit(fl, xx[mask], np.log(noisy_y[mask]), [1.,1.,1.])
```

```python
noisy_y[noisy_y > 0]
```

```python

%matplotlib notebook
plt.plot(xx,fl(xx,*popt),'-')
plt.plot(xx[mask], np.log(noisy_y[mask]), 'o--')
plt.plot(xx,np.log(yy),'--')
```
