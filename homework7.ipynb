# Восстановление функции

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x,a,b,c): #наша функция
    arg=1+((x-a)/b)**2 
    return c/arg
```

```python
xa, xb= -2.,2.
npts=20
```

```python
xx = np.linspace(xa,xb)
```

```python
np.random.seed(123)
```

```python
noise=np.random.normal(loc=0, scale=0.03, size=xx.size) #создаем шумы
```

```python
a=0.5
b=1
c=2
```

```python
yy = f(xx, a, b, c)
```

```python
noisy_y=yy + 4*noise #добавляем шумы
%matplotlib notebook
plt.plot(xx,yy,'--', label=r'Первоначальная функция')
plt.plot(xx, noisy_y, 'o', label=r'Зашумленные данные')
plt.legend(loc='best')
```

```python
from scipy.optimize import curve_fit
```

```python
popt, pcov = curve_fit(f, xx, (noisy_y), [1.,1.,1.]) #восстанавливаем функцию
```

```python
%matplotlib notebook
plt.plot(xx,(f(xx,*popt)),'-', label=r'Восстановленная функция')
plt.plot(xx, (noisy_y), 'o', label=r'Зашумленные данные')
plt.plot(xx,(yy),'--', label=r'Первоначальная функция')
plt.legend(loc='best')
```
