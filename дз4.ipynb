# Дифференцирование

Односторонняя двухточечная система:

```python
def diff(f,a,h):
    '''Производная: f-функция, a-значение х, h-эпсилон'''
    return (f(a+h)-f(a))/(h)
diff(lambda x: (x**11), 1.2, 0.0000001)
```

Центральная двухточечная система

```python
def diff2(f,a,h):
    '''Производная: f-функция, a-значение х, h-эпсилон'''
    return (f(a+h)-f(a-h))/(2*h)
diff2(lambda x: (x**11), 1.2, 0.0000001)
```

```python
def func(x): 
    '''функция 11*х**10'''
    return 11*x**10    
```

```python
import matplotlib.pyplot as plt
import numpy as np
def plot_all(x, n, k):
    '''Построение графиков (x-фиксированное значение функции)'''
    xx = [10 ** (k - i / 10) for i in range (n)]
    y1 = [abs(diff(lambda x: (x**11), x, i) - func(x)) for i in xx]
    y2 = [abs(diff2(lambda x: (x**11), x, i) - func(x)) for i in xx]
    plt.grid(True)
    plt.loglog(xx, y1, 'or--', label=u'Односторонняя' )
    plt.loglog(xx, y2, 'og--', label = u'Центральная')
```

```python
plot_all(1.2, 250, 1)
plt.grid(True)
plt.legend()
plt.show()
```

```python
plot_all(1.2, 100, 1)
plt.grid(True)
plt.legend()
plt.show()
```

При h < 10^-16 значения становятся равны константе, т.к. производная с h меньше чем машинный эпсилон обнуляется. 
Флуктуации - результат неточного округления и накопления ошибки при малых значениях приращения аргумента. Видно, что центральная производная дает более точное приближение, т.к. минимум разности у нее меньше.  От h ≈ 10^0 оба способа дают примерно одинаковый плохой результат.
