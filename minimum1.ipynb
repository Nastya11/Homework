# Нахождение минимума функции

```python
import math
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
def f(x): 
    '''наша функция -cosx'''
    f=-(math.cos(x))
    return f
```

```python
def minim(a,b,eps):
    '''ищем минимум функции методом золотого сечения'''
    F=(1+math.sqrt(5))/2 #пропорция золотого сечения
    print ("а начальное = ", a, '; b начальное =', b, '\n')
    while(abs(b-a)>eps):
        x1=b-(b-a)/F
        x2=a+(b-a)/F
        y1=f(x1)
        y2=f(x2)
        if (y1>=y2):
            a=x1
        else:
            b=x2
        print('a =',a,' b =',b) #вывожу a,b, чтобы видеть, как работает метод
    x=(a+b)/2
    print ('\n result: \nx min = ', x)
    print ('y min = ', f(x))
    return x
```

```python
a = 4.5
b = 6.5
```

Найдем производную: (-cosx)'=sinx. Условие экстремума: f'(x)=0 => x=Пn, n принадлежит целым числам. Возьмем промежуток, где находится 1 минимум, например, [4.5,6.5]. Минимум будет в точке х=2П = 6,2831853... Посмотрим, какой ответ получится нашим методом:

```python
%matplotlib notebook
xx = np.arange(a - 0.5, b + 0.5, 0.002)
minimum_1 = minim(a, b, 10 ** (-8))
plt.plot(xx, [f(el) for el in xx], '-')
plt.plot(xx, [f(minimum_1) for el in xx], '-')
```
